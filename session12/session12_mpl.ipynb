## Matplotlib

### 1. Introduction

Matplotlib is a python library for two-dimensional graphics created in 2003 as a free alternative to the proprietary MATLAB. The project was so successful that now matplotlib is the standard for two-dimensional graphics in data analysis in python. Using matplotlib you can draw very complex graphics, so the threshold of understanding is high. For simple graphs, there are convenient wrappers, but customization requires an understanding of the details.

```python
import matplotlib as mpl  # common alias for matplotlib
```

Matplotlib consists of many modules and classes. Drawing in matplotlib consists of sequentially arranging objects on a canvas. Objects form a hierarchical structure. The user's task is to describe this structure, and the library itself will be engaged in drawing.

The drawing in matplotlib has the following structure. There is a drawing object (Figure), it has a canvas (Canvas) on which there can be 1 or more drawing areas (Axes). Drawing areas contain drawing elements (Artists). Regular users spend most of their time working with drawing elements by arranging them in the right order on the drawing areas. Elements of the picture (Artists) can be divided into two large groups:

 - **Primitives** - the simplest elements of a picture, such as lines, rectangles, text, images, etc.
 - **Containers** - drawing (Figure), drawing area (Axes), coordinate axes (Axis), marks on axes (Ticks).  

The library manipulates very simple graphic primitives (line, point, segment, polygon), which is completely inconvenient for people - they think in high-level abstractions. Therefore, in matplotlib there are two wrappers - pylab and pyplot. The first is considered an unsuccessful experiment - do not use it. We will work with pyplot.

```python
import matplotlib.pyplot as plt  # common alias for pyplot
```

### 2.2. Complex examples

#### 2.2.1. Working with texts

Let's look at an example showing what elements of the picture can be on the drawing area and what settings they can have.


We will draw a picture with:
 - one drawing area
 - figure title
 - plot title
 - graph of $\cos(x^2)$
 - annotation
 - text in a box
 - X and Y labels
 - legend
 - grid
 
All kwargs are optional. You can delete them to see how these items are displayed by default.

```python
#  Data generation
step = 0.01

# create array of numbers from 0 to 2pi with step of 0.01
x = np.arange(0.0, 2*np.pi + step, step)
#  calculate cos(x^2)
y = np.cos(x * x)

# Create figure 
fig = plt.figure(
    figsize=(16, 9),  # figure size in inches
    facecolor='whitesmoke',  # background color
    dpi=200  # 200 dots per inch
)

# Figure title with 30th font size
plt.suptitle('This is a title for the whole figure', fontsize=30)

# Plot title
plt.title('This is a title for our plot', fontsize=20)

# Drawing cos(x^2)
plt.plot(
    x, y,
    color='red',  # Red color
    linestyle='--',  # Dashed style
    alpha=0.4,  # (1 - transparency)
    linewidth=10,  # Width
    label='This is a legend for graph of $\cos(x^2)$' # Legend. You can use TeX!
)

# Annotation
plt.annotate(
    'This is an annotation\n for $\cos(x^2)$ with fancy arrow\n pointing at grpah',
    xy=(1.15, -0.25),  #  where arrow will be pointed at
    arrowprops=dict(arrowstyle='fancy'),
    xytext=(0, -0.5)  # text location
)

# Place legend
plt.legend(
    loc='lower left',
    borderaxespad=5 
)

# X and Y labels
plt.xlabel(
    'This is an X axis label',  # Text
    fontdict=dict(family='serif', color='darkred', weight='normal', size=16)  # Font properties
)
plt.ylabel(
    'This is a Y axis label',
    fontdict=dict(family='monospace', color='peru', weight='light', size=25)
)

# Add text
plt.text(
    2.0, 0.2,  # location of text's beginning 
    'This is a custom text',  # text
    fontsize=26,  
    color='firebrick',  # color for letters
    rotation=20,
    bbox=dict(  # frame properties
        alpha=0.4, # (1-transparency)
        edgecolor='slateblue',  # frame color
        facecolor='azure',  # background color
        pad=1., # frame padding
        boxstyle='round'  # frame form
    )
)

# Add grid
# vertical lines are Xticks, horizontal - Yticks
plt.grid(True)

plt.show()
del step, x, y, fig
```

#### 2.2.2. Polar coordinates

The drawing area may be non-rectangular, for example if you use polar coordinates. In this case, you will have to create your own drawing area. We will create it using the add_axes method by specifying the dimensions of the rectangle in which the drawing area will be located. Here it is worth saying that in the figure the coordinates are relative, in the range from 0 to 1. In this example, we will use a more object-oriented approach and will call methods of objects, not functions of the plt module. Using the functions of the plt module is considered a bad manner since you constantly operate with a global state to which you do not even have access.

```python
# Prepare data
step = 0.01
phi = np.arange(0.0, 2*np.pi + step, step)
rho1 = np.cos(8 * phi)
rho2 = np.sin(4 * phi)

fig = plt.figure(figsize=(16, 9))

# lower left corner in  (0,0)
# Width = 1, Height = 1 (i.e. the whole figure)
ax = fig.add_axes([0,0,1,1], polar=True, facecolor='ivory')

ax.set_title('Polar plots', fontsize=20)

ax.plot(
    phi, rho1,
    color='orange', 
    marker='o',  
    markersize=3, 
    label='$\cos(8x)$'
)

ax.plot(
    phi, rho2,
    linewidth=5,
    color='olive',
    label='$\sin(4x)$'
)

# Placing legend
ax.legend(
    bbox_to_anchor=(1.2, 0.54), #custom position
    fancybox=True,
    shadow=True
)

ax.grid(True)

plt.show()
del step, phi, rho1, rho2, fig, ax
```

#### 2.2.3. Different types of graphs in one picture

We will draw:
 - a set of random measurements of $cos(x)$ with white noise of zero mean and 0.25 dispersion
 - graph of cos(x)
 - and the area of errors within 1 sigma

```python
# Prepare data
scale = 0.5
n_points = 100
x = 2 * np.pi * np.sort(np.random.rand(100))
y = np.cos(x)
y_noise = y + np.random.normal(scale=0.25, size=n_points)


fig = plt.figure(figsize=(16, 9))

ax = fig.add_axes([0, 0, 1, 1])  

ax.scatter(
    x, y_noise,
    marker='^',  
    s=50,  
    color='red'  
)

# cos(x)
ax.plot(
    x, y,
    color='grey',
    linestyle='--',
    alpha=0.5,
    linewidth=4
)

# 1 sigma deviations
for shift in (scale, -scale):
    ax.plot(
        x, y + shift,
        alpha=0.5,
        color='darkgrey',
        linewidth=6
    )

# Fill between -sigma and sigma
ax.fill_between(
    x, y - scale, y + scale,
    alpha=0.5,
    color='lightgrey',
)

# labels
ax.set_xlabel('Uniformly random input: x', fontsize=15)
ax.set_ylabel('Readout: $y=cos(x)+\mathbf{noise}$', fontsize=15)
    
ax.grid(True)

plt.show()
del scale, n_points, x, y, y_noise, fig, ax
```

As you saw in the last examples, most of the painting comes through working with the drawing area (Axes). This container object contains other containers and primitive objects. It can be seen that there are many methods for creating and changing the properties of elements of drawings and graphic primitives. Everything that can be done through pyplot can be done working directly with axes instances. The following matches are obtained:

Method of Axes object -> Description -> object of Artist class created by the method -> Container with created object

    ax.annotate -> annotation -> Annotate -> ax.texts

    ax.text -> text -> Text -> ax.texts

    ax.bar -> bar chart -> Rectangle -> ax.patches

    ax.errorbar -> mustache chart -> Line2D & Rectangle -> ax.lines и ax.patches

    ax.fill -> fill -> Polygon -> ax.patches

    ax.hist -> bar chart -> Rectangle -> ax.patches

    ax.imshow -> image -> AxesImage -> ax.images

    ax.legend -> legend -> Legend -> ax.legends

    ax.plot -> XY graph -> Line2D -> ax.lines

    ax.scatter -> dot plot -> PolygonCollection -> ax.collections


#### 2.2.4. Several axes

It is often necessary to arrange multiple areas on the same chart. To do this, create several drawing areas and work with each separately. In the following example, we will plot the graph of the function $x^2(2+\sin(1/x))$ and add another drawing area for the most interesting part of the graph

```python
def crazy_func(x):
    return x * x * (2+np.sin(1/x))

fig = plt.figure(figsize=(16, 9))

fig.suptitle('A crazy function example')

# Main graph
big_ax = fig.add_axes([0, 0, 1, 0.9])
big_ax.plot(
    np.linspace(0.001, 1, 1000),
    np.apply_along_axis(crazy_func, 0, np.linspace(0.001, 1, 1000)),
    label='$x^2(2+\sin(1/x))$'
)
big_ax.text(
    0, 2.5,
    '''
    This function is differentiable at 0, has global minimum there and 
    infinitely many local minima and maxima in the neighbourhood
    '''
)

big_ax.legend(bbox_to_anchor=(0.2, 0.54))
big_ax.grid(True)

# Additional graph
small_ax = fig.add_axes([0.5, 0.1, 0.4, 0.6])
small_ax.plot(
    np.linspace(0.001, 0.1, 1000),
    np.apply_along_axis(crazy_func, 0, np.linspace(0.001, 0.1, 1000))
)
small_ax.grid(True)


plt.show()
del crazy_func, fig, big_ax, small_ax
```

#### 2.2.5. Multiple windows

In the previous example, we created two drawing areas, but their location had to be set explicitly using coordinates. In most cases, this exact level of customization is not needed. For this, multi-windows exist in matplotlib. In many ways, they are similar to the drawing areas. You can create them from the Figure object using the **add_subplots** method.

When using the add_subplots method, a picture is split like a table into a number of rows and columns. The resulting cells are not drawn anywhere - this is a convenient abstraction. Next, you can select one of these cells, and draw in it. To do this, pass three numbers or a three-digit number to the subplots function. The first digit indicates the number of rows, the second - columns, the third - the serial number of the cell in which you want to draw (the numbering starts from 1 goes from left to right and from top to bottom). For example, calling fig.add_subplots (3, 3, 7) will break the picture into 9 equal parts and return you a cell from the lower left corner.

The advantage of the approach is that you get the same size drawing areas without thinking about what coordinates they should have.

```python

t = np.linspace(0, 2*np.pi, 1000)


fig = plt.figure(figsize=(16, 9))
fig.suptitle('Lissajous figures')

nrows = 4
ncols = 4

for a in range(1, nrows + 1):
    for b in range(1, ncols + 1):
        # Choose cell
        ax = fig.add_subplot(nrows, ncols, (a - 1) * ncols + b)
        # Draw a Lissajous curve
        ax.plot(
            np.cos(a*t),
            np.sin(b*t),
        )
        # Remove axis
        ax.set_axis_off()
        # Add title
        ax.set_title('a={}, b={}'.format(a, b))

plt.show()
```

More general way:

```python
t = np.linspace(0, 2*np.pi, 1000)

figure, axes = plt.subplots(4,4, figsize=(16,9))

nrows = 4
ncols = 4

for a in range(nrows):
    for b in range(ncols):
        # Выбираем ячейку
        ax = axes[a][b]
        # Рисуем кривую Лиссажу
        ax.plot(
            np.cos((a+1)*t),
            np.sin((b+1)*t),
        )
        # Обираем оси координат (они только засоряют рисунок)
        ax.set_axis_off()
        # Делаем подписи
        ax.set_title('a={}, b={}'.format(a+1, b+1))

plt.show()
```

Sometimes you want to collect several graphs in one picture, but their sizes should be different. To do this, use the **GridSpec** object.

```python
from matplotlib import gridspec  


fig = plt.figure(figsize=(16, 9))
fig.suptitle('Zoo of hyperbolic functions', fontsize=20)

# Create "spec"
spec = gridspec.GridSpec(nrows=3, ncols=4)

fig.add_subplot(spec[0, :3])  # first three cells of first row
fig.add_subplot(spec[:2, 3])
fig.add_subplot(spec[2, 1:])
fig.add_subplot(spec[1:, 0])
fig.add_subplot(spec[1, 1:3])

trig_funcs = [np.sinh, np.tanh, np.cosh, np.arcsinh, np.arccosh]
points = np.linspace(-2, 2, 100)

for ax, func in zip(fig.axes, trig_funcs):
    ax.plot(
        points,
        func(points),
        label='{}(x)'.format(func.__name__)
    )
    # Place legend at "the best" possible position
    ax.legend(loc='best')
    ax.grid(True)

plt.show()
del fig, spec, ax, func
```

#### 2.2.6. Coordinate axis

In matplotlib, coordinate axis are a drawing element, or rather a container of type Artist. Examples are the $x$, $y$ axis in the Cartesian coordinate system, the $r$, $\phi$ axis in the polar coordinate system. Each axis contains three groups of graphic primitives: tick marks, tick labels and axis labels.

Major and minor ticks are distinguished. To specify exact type you can use **which** argument in **get_ticks** method. By default only major ticks are displayed.

```python
fig = plt.figure(figsize=(16, 9))

ax = fig.add_subplot(111)
ax.set_title('Process of studying matplotlib', fontsize=25, color='cadetblue')

ax.plot(
    np.arange(70),
    2 - np.exp(0.01 * np.arange(70)),
    color='teal'
)

ax.set_xticks(np.arange(5, 85, 10), minor=True)
ax.set_xticks(np.arange(0, 80, 10))

ax.set_xlim(0, 80)


ax.set_xticklabels(
    ['{} mins'.format(t) for t in range(0, 80, 10)],
    rotation=45,  
    fontsize=15,
    color='cadetblue'
)

ax.tick_params(
    axis='x',  
    which='major',
    direction='inout', # ticks will intersect x-axis
    length=20,  
    width=5, 
    color='powderblue',
    top=False, bottom=True,  # bottom only
    labeltop=False, labelbottom=True  # bottom only for labels too
)

ax.set_yticks(np.arange(0.1, 1, 0.2), minor=True)
ax.set_yticks(np.arange(0, 1.1, 0.2))
ax.set_ylim(-0.2, 1.2)
ax.set_yticklabels(
    ['Leave me alone', 'I don\'t give a damn', 'Indifference', 'Interest', 'Inquisitiveness', 'Pure excitement'],
    fontsize=15,
    color='cadetblue'
)
ax.tick_params(
    axis='y', which='major',
    direction='in',
    length=20, width=5, color='powderblue',
    left=True, right=True,
    labelleft=True, labelright=True
)

# Axis labels
ax.set_xlabel('Duration', color='teal', fontsize=20)
ax.set_ylabel('Motivation', color='teal', fontsize=20)

# Dashed lines for major ticks
ax.grid(True, axis='x', which='major', color='dimgrey', linestyle='--')
# Solid lines for minor ticks
ax.grid(True, axis='x', which='minor', color='lightgrey', linestyle=':')

# it's all the same for Y
ax.grid(True, axis='y', which='major', color='grey', linestyle='-.')
ax.grid(True, axis='y', which='minor', color='lightgrey', linestyle='-')

plt.show()
```
